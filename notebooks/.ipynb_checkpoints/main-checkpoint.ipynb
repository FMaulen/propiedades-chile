# Importar librerias

```python
import pandas as pd
import os
from os.path import basename
```

```python
def relative_path():
    path = '.'
    path = os.path.realpath('.')
    return path[:-len(basename(path))]

path = os.path.join(relative_path(), 'data\\raw\\2023-03-08 Precios Casas RM.csv')
```

# Carga y copia del dataset

```python
data = pd.read_csv(path)
data = data.copy()
```
